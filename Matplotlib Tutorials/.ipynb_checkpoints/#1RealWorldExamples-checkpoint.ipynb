```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
```

### Real World Examples

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(10,5))


plt.title('Gas Prices Over Time (in USD)')

# plt.plot(gas.Year, gas.Australia, 'r.-', label="Australia")
# plt.plot(gas.Year, gas.Canada, 'b.-', label="Canada")
# plt.plot(gas.Year, gas.Japan, 'g.-', label = "Japan")


print(gas.Year[::3])

countries_to_look_at = ['France', 'Italy', 'Australia', 'Canada']
for Country in gas.columns[1:]:
    if Country in countries_to_look_at:
        plt.plot(gas.Year, gas[Country], marker='.', label = Country)
        

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('USD')

plt.legend()

timestamp = datetime.datetime.now().strftime("%m_%d_%Y_%H%M%S")
plt.savefig(f'GasPricesOverTime Analysis_{timestamp}.png', dpi = 180)

plt.show()
plt.close()
```

### Pie Charts

```python
fifa = pd.read_csv('fifa_data.csv')

#Get the list of all the columns in fifa dataset
columns = fifa.columns.tolist()

#Ignoring the first column name
columns_list = columns[1:]

fifa_columns_dict = {index: column_name for index, column_name in enumerate(columns_list)}
print(fifa_columns_dict)
print("\n\n")

#Checking and getting the index of specific column name
word = 'Preferred Foot'
if word in columns_list:
    index = columns_list.index(word)
    print(f'{word} found at {index}')
else:
    print(f'{word} not found at {columns_list}')
```

```python
#Accessing Preferred Foot value for 

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct = '%.2f')
plt.title('Foot Preference of FIFA Players')
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')


#Both are logical but below is more professional comparitively
# fifa['Weight'] = fifa['Weight'].str.replace('lbs','', regex = False).astype(float)
# fifaWeight = fifa['Weight'];


fifa.Weight = [x.strip('lbs') if type(x)==str else x for x in fifa.Weight]
fifa.Weight = fifa.Weight.astype(float)      #We are converting into numeric, as we are compoaring with numbers in logic.


light = fifa.loc[fifa.Weight < 125].count().iloc[0]
lighty_medium = fifa.loc[(fifa.Weight >= 125) &(fifa.Weight > 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150)&(fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175 )&(fifa.Weight < 200)].count().iloc[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count().iloc[0]

weights = [light, lighty_medium, medium, medium_heavy, heavy]
labels = ['Under 125','125 -150','150-175','175-200','Over 200']

plt.title('Weight distribution of FIFA Players (in lbs)')
plt.pie(weights, labels = labels, autopct = '%.2f%%')
plt.show()
```

### Box plots

```python
plt.style.use('default')
plt.figure(figsize=(5,8))

Barcelona = fifa.loc[fifa.Club == 'FC Barcelona' ]['Overall']
Manchester = fifa.loc[fifa.Club == 'Manchester United']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Barcelona', 'Manchester', 'NE Revolution']

plt.title('Professional soccer Team Comparison')
plt.boxplot([Barcelona, Manchester, revs], labels = labels)
plt.show()
```
